```python
from idanalysis.analysis import KickmapAnalysis
import numpy as np
import matplotlib.pyplot as plt
```

```python
kmap_fname = './kickmap-ID_curr_250.txt'
kick_anl = KickmapAnalysis(kmap_fname)
```

```python
plt.figure()

rx0, ry0, pxf, pyf, rxf, ryf = kick_anl.get_kicks_at_plane(
    indep_var="x", plane=1.7e-3)
plt.plot(1e3*rx0, 1e6*pxf, '-o', label='kick x')
plt.plot(1e3*rx0, 1e6*pyf, '-o', label='kick y')

plt.xlabel('rx [mm]')
plt.ylabel('kick [urad]')
plt.legend()
```
